```python
import os
import pandas
import numpy as np
```

```python
folder = './result/'
files = os.listdir(folder)

max_withsin = []
max_woutsin = []
for dir_name in files:
    if dir_name[0] != '.':
        path = './result/' + dir_name + '/'
        file = ['with_singan.txt', 'without_singan.txt']
        f = open(path + file[0], 'r')
        lines = f.readlines()
        withsin_accuracy = []
        woutsin_accuracy = []
        for line in lines:
             test_acc = float(line.split(',')[2].split(':')[1])
             #print(train_acc)
             withsin_accuracy.append(test_acc)
        f.close()
        
        #print('withsin', max(withsin_accuracy))
        
        f = open(path + file[1], 'r')
        lines = f.readlines()
        for line in lines:
             test_acc = float(line.split(',')[2].split(':')[1])
             #print(train_acc)
             woutsin_accuracy.append(test_acc)
        f.close()
        #print('woutsin', max(woutsin_accuracy))
        max_withsin.append(max(withsin_accuracy))
        max_woutsin.append(max(woutsin_accuracy))
        
print('mean + std of accuracy with SinGan Data Augumentation: {} + {}'.format(np.mean(max_withsin), np.std(max_withsin)))
print('mean + std of accuracy without SinGan Data Augumentation: {} + {}'.format(np.mean(max_woutsin), np.std(max_woutsin)))

        
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
plt.plot(max_withsin)
plt.plot(max_woutsin)
plt.title('Highest Test Accuracy Comparison Over With-singan and Without-singan')
plt.xlabel('experiment')
plt.ylabel('Test Accuracy')

plt.legend(['with SinGan', 'wout SinGan'])
plt.savefig('test accuracy.png')
plt.show()
```
